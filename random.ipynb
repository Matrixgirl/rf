```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv(r'C:\random1\Grp1.csv')
```

```python
data
```

```python
data['Approved'] = data['Approved'].map({'Yes': 1, 'No': 0})
```

```python
data
```

```python
label_encoders = {}
for column in data.columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le
```

```python
X = data[['Age', 'Income ($)', 'Credit_Score', 'Loan_Amount ($)']]
y = data['Approved']
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)
```

```python
new = pd.DataFrame({
    'Age': [29],
    'Income ($)': [37000],
    'Credit_Score': [675],
    'Loan_Amount ($)': [7000]
})
```

```python
prediction = model.predict(new)
predicted_label = label_encoders['Approved'].inverse_transform(prediction)
```

```python
print("Prediction for new applicant → Approved?:", predicted_label[0])
```

```python
import joblib

# Assuming `model` is your trained Random Forest model
# and `label_encoders` is your dictionary of label encoders

# Save the trained model
joblib.dump(model, 'loan_approval_rf_model.joblib')

# Save the label encoders
joblib.dump(label_encoders, 'loan_approval_label_encoders.joblib')

print("Model and label encoders saved successfully!")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Actual and predicted labels
y_true = ['Yes', 'No', 'Yes', 'No', 'Yes']
y_pred = ['Yes', 'Yes', 'Yes', 'No', 'No']

# Convert to binary labels (1 = Yes, 0 = No)
y_true_bin = [1 if label == 'Yes' else 0 for label in y_true]
y_pred_bin = [1 if label == 'Yes' else 0 for label in y_pred]

# Calculate metrics
accuracy = accuracy_score(y_true_bin, y_pred_bin)
precision = precision_score(y_true_bin, y_pred_bin)
recall = recall_score(y_true_bin, y_pred_bin)
f1 = f1_score(y_true_bin, y_pred_bin)
conf_matrix = confusion_matrix(y_true_bin, y_pred_bin)

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
print("Confusion Matrix:\n", conf_matrix)
```
